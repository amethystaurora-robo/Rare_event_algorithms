```python
import pandas as pd
import numpy as np
import os
import zipfile
```

```python
dfs = []
for file in os.listdir("."):
    if file.startswith("data_") and file.endswith(".csv.zip"):
        df = pd.read_csv(file, compression='zip')
        dfs.append(df)

#combined_df = pd.concat(dfs, ignore_index=True)
```

```python
#load and concat data transition data
salts=['amocs/salt_40y_200traj_7000k.csv','amocs/salt_40y_200traj_8000k.csv','amocs/salt_40y_200traj_9000k.csv',
           'amocs/salt_40y_1000traj_7000k.csv','amocs/salt_40y_1000traj_8000k.csv','amocs/salt_40y_1000traj_9000k.csv']
amocs=['amocs/amoc_trans_off_40y_200traj_0k.csv','amocs/amoc_trans_off1st_40y_200traj_-1000k.csv',
       'amocs/amoc_trans_off1st_40y_200traj_-2000k.csv','amocs/amoc_trans_off1st_40y_1000traj_-500k.csv',
      'amocs/amoc_trans_off1st_40y_1000traj_-1000k.csv','amocs/amoc_trans_off1st_40y_1000traj_-2000k.csv']

zip_path ="/Users/amethyst/Documents/Rare_event_algorithms/amocs.zip"
trans = []

with zipfile.ZipFile(zip_path) as z:
    print(z.namelist())  # 👈 run this once to see actual paths

    for salt in salts:
        with z.open(salt) as f:
            df = pd.read_csv(f)
            df['var'] = 'salt'
            df['Transition'] = 'Y'
            trans.append(df)

    for amoc in amocs:
        with z.open(amoc) as f:
            df = pd.read_csv(f)
            df['var'] = 'AMOC'
            df['Transition'] = 'Y'
            trans.append(df)
```

```python
#combine transitioned with normal data
final_df = pd.concat(trans + dfs, ignore_index=True)
```
